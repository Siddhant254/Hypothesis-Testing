## 2-Sample t-test  - two tail test

```python
from scipy.stats import ttest_ind
import numpy as np

#week 1 data
week1 = np.random.uniform(low=10.0 , high=25.3 , size=(50,))

#week 2 data
week2 = np.random.uniform(low=10.2 , high=26.5 , size=(50,))

print("\nweek1 data : -\n")
print(week1)
print("\nweek2 data : -\n")
print(week2)

#Mean of week1 data
week1_mean = np.mean(week1)
print("\nweek1 mean value :",week1_mean)

#Mean of week2 data
week2_mean = np.mean(week2)
print("\nweek2 mean value :",week2_mean)

#Standard deviation of week1 data
week1_std = np.std(week1)
print("\nweek1 std value :",week1_std)

#Mean of week2 data
week2_std = np.std(week2)
print("\nweek2 std value :",week2_std)

#t-test
t_calc , pval = ttest_ind(week1,week2)   #2-TAIL TEST
print("\nP-Value",pval)

if pval<0.05:  #alpha=5%
    print("We will be rejecting null hypothesis")
    
else:
    print("We have failed to reject null hypothesis")

```

## 2-sample t-test - right tail

```python
from scipy.stats import ttest_ind
import numpy as np

#week 1 data
week1 = np.random.uniform(low=10.0 , high=25.3 , size=(50,))

#week 2 data
week2 = np.random.uniform(low=10.2 , high=26.5 , size=(50,))

print("\nweek1 data : -\n")
print(week1)
print("\nweek2 data : -\n")
print(week2)

#Mean of week1 data
week1_mean = np.mean(week1)
print("\nweek1 mean value :",week1_mean)

#Mean of week2 data
week2_mean = np.mean(week2)
print("\nweek2 mean value :",week2_mean)

#Standard deviation of week1 data
week1_std = np.std(week1)
print("\nweek1 std value :",week1_std)

#Mean of week2 data
week2_std = np.std(week2)
print("\nweek2 std value :",week2_std)

#t-test
t_calc , pval = ttest_ind(week1,week2,alternative='greater')   # RIGHT TAIL TEST
print("\nP-Value",pval)

if pval<0.05:  #alpha=5%
    print("We will be rejecting null hypothesis")
    
else:
    print("We have failed to reject null hypothesis")

```

## 2-SAMPLE T-TEST - LEFT TAIL

```python
from scipy.stats import ttest_ind
import numpy as np

#week 1 data
week1 = np.random.uniform(low=10.0 , high=25.3 , size=(50,))

#week 2 data
week2 = np.random.uniform(low=10.2 , high=26.5 , size=(50,))

print("\nweek1 data : -\n")
print(week1)
print("\nweek2 data : -\n")
print(week2)

#Mean of week1 data
week1_mean = np.mean(week1)
print("\nweek1 mean value :",week1_mean)

#Mean of week2 data
week2_mean = np.mean(week2)
print("\nweek2 mean value :",week2_mean)

#Standard deviation of week1 data
week1_std = np.std(week1)
print("\nweek1 std value :",week1_std)

#Mean of week2 data
week2_std = np.std(week2)
print("\nweek2 std value :",week2_std)

#t-test
t_calc , pval = ttest_ind(week1,week2,alternative='less')   # LEFT TAIL TEST
print("\nP-Value",pval)

if pval<0.05:  #alpha=5%
    print("We will be rejecting null hypothesis")
    
else:
    print("We have failed to reject null hypothesis")

```
